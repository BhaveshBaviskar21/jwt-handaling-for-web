# jwt deployed and stored in cookie

```python
def send_request(payload):
    url1 = "http://localhost:21066/api/login"
    url2 = "http://localhost:21066/api/quote"
    try:
        start_time = time.time()
        response = requests.post(url1, json=payload)
        end_time = time.time()
        runtime1 = end_time - start_time
        start_time = time.time()
        res = requests.get(url2, headers={"Cookie":'token='+response.cookies['token']})
        end_time = time.time()
        runtime2 = end_time - start_time
        return runtime1, runtime2, res.status_code
    except Exception as e:
        return None, str(e)
```

```python
credentials = [
    {"email": "admin1@gmail.com", "password": "123456"},
    {"email": "admin2@gmail.com", "password": "123456"},
    {"email": "admin3@gmail.com", "password": "123456"},
    {"email": "admin4@gmail.com", "password": "123456"},
    {"email": "admin5@gmail.com", "password": "123456"},
    {"email": "admin6@gmail.com", "password": "123456"},
    {"email": "admin7@gmail.com", "password": "123456"},
    {"email": "admin8@gmail.com", "password": "123456"},
    {"email": "admin9@gmail.com", "password": "123456"},
    {"email": "admin10@gmail.com", "password": "123456"},
]
```

```python
results = []
results1 = []
```

```python
num_requests = 100

worker_credentials = credentials * (num_requests // len(credentials))
worker_credentials += credentials[: num_requests % len(credentials)]

with ThreadPoolExecutor(max_workers=10) as executor:
    futures = [executor.submit(send_request,worker_credentials[i]) for i in range(num_requests)]
    for future in futures:
        try:
            runtime1,runtime2, status_code = future.result()
            results.append((runtime1, status_code))
            results1.append((runtime2, status_code))
        except Exception as e:
            results.append((None, str(e)))
            results1.append((None, str(e)))
```
